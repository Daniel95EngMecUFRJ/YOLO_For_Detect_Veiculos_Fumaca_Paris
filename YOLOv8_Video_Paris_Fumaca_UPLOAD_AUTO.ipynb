# YOLOv8 em Vídeo de Paris com Detecção Automática de Pasta de Saída

Este notebook permite:
- Upload de vídeo com carros e fumaça
- Processamento com YOLOv8
- Contagem por frame
- Detecção automática da pasta `runs/detect/*`
- Link de download do vídeo processado

```python
# 1. Instalar Ultralytics
!pip install ultralytics --upgrade -q
from ultralytics import YOLO
```

```python
# 2. Upload do vídeo
from google.colab import files
uploaded = files.upload()
```

```python
# 3. Rodar YOLOv8
video_path = list(uploaded.keys())[0]
model = YOLO('yolov8n.pt')
results = model.predict(source=video_path, save=True, conf=0.25)
```

```python
# 4. Descobrir automaticamente o nome da pasta de saída
import os
from pathlib import Path

base_dir = Path('runs/detect')
output_dirs = sorted(base_dir.glob('predict*'), key=os.path.getmtime, reverse=True)
if output_dirs:
    latest_dir = output_dirs[0]
    print("📂 Pasta de saída YOLO:", latest_dir)
    for file in os.listdir(latest_dir):
        if file.endswith('.mp4'):
            print("🎥 Vídeo com detecções salvo em:", latest_dir / file)
else:
    print("❌ Nenhuma pasta detectada em runs/detect")
```

```python
# 5. Contagem de objetos por frame
from collections import Counter
for i, result in enumerate(results):
    classes = result.boxes.cls.cpu().numpy()
    labels = [model.names[int(c)] for c in classes]
    contagem = Counter(labels)
    print(f"Frame {i + 1}: {contagem}")
```
