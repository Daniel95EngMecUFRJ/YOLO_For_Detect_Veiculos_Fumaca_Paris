# Detecção de Veículos e Fumaça com YOLOv8

Este notebook aplica o YOLOv8 pré-treinado para detectar veículos e visualizar emissão de fumaça em imagens urbanas, com foco em aplicações ambientais para cidades sustentáveis como Paris.

```python
# 1. Instalar o pacote Ultralytics
!pip install ultralytics --upgrade -q
from ultralytics import YOLO
```

```python
# 2. Fazer upload da imagem
from google.colab import files
uploaded = files.upload()
```

```python
# 3. Carregar modelo YOLOv8 pré-treinado
model = YOLO('yolov8n.pt')  # Modelo leve para teste rápido
# Mostrar as classes que ele reconhece
print(model.names)
```

```python
# 4. Rodar detecção na imagem enviada
results = model.predict(source=list(uploaded.keys())[0], save=True, conf=0.25)
results[0].show()
```

⚠️ Neste ponto, você verá os veículos detectados. A fumaça **ainda não é reconhecida automaticamente** porque não faz parte das classes padrão do COCO. Mas podemos **segmentar manualmente** como 'áreas de interesse' ou treinar um novo modelo com imagens que tenham a classe 'fumaça'.
